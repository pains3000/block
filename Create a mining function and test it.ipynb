```python
import hashlib

def sha256(message):
    return hashlib.sha256(message.encode('ascii')).hexdigest()

def mine(message, difficulty=1):
    assert difficulty >= 1
    prefix = '0' * difficulty  # Change '1' to '0' to match with leading zeros
    print("prefix", prefix)
    for i in range(1000):
        digest = sha256(message + str(i))  # Concatenate message and nonce directly
        print("testing => " + digest)
        if digest.startswith(prefix):
            print("after " + str(i) + " iterations found nonce: " + str(i))
            return i  # i= nonce value

mine("test message", 2)
```
